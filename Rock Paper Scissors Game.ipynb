```python
# Import the random module to generate random numbers
import random

# Print the instructions and rules of the game
print('Winning rules of the game ROCK PAPER SCISSORS are :\n'
      + "Rock vs Paper -> Paper wins \n"
      + "Rock vs Scissors -> Rock wins \n"
      + "Paper vs Scissors -> Scissors wins \n")

# Create an infinite loop to allow the game to be played multiple times
while True:
    # Prompt the player to enter their choice (1 for Rock, 2 for Paper, 3 for Scissors)
    print("Enter your choice \n 1 - Rock \n 2 - Paper \n 3 - Scissors \n")
    
    # Take the input from the player and convert it to an integer
    choice = int(input("Enter your choice: "))

    # Check if the player's choice is valid (between 1 and 3)
    while choice > 3 or choice < 1:
        choice = int(input('Enter a valid choice, please: '))

    # Determine and display the player's choice as 'Rock,' 'Paper,' or 'Scissors'
    if choice == 1:
        choice_name = 'Rock'
    elif choice == 2:
        choice_name = 'Paper'
    else:
        choice_name = 'Scissors'

    print('User choice is \n', choice_name)
    print("Now it's the system's turn....")

    # Generate a random choice for the system (1 for Rock, 2 for Paper, 3 for Scissors)
    system_choice = random.randint(1, 3)

    # Ensure the system's choice is different from the player's choice
    while system_choice == choice:
        system_choice = random.randint(1, 3)

    # Determine and display the systems choice as 'Rock,' 'Paper,' or 'Scissors'
    if system_choice == 1:
        system_choice_name = 'Rock'
    elif system_choice == 2:
        system_choice_name = 'Paper'
    else:
        system_choice_name = 'Scissors'

    print("System's choice is \n", system_choice_name)
    print(choice_name, 'Vs', system_choice_name)

    # Determine the result of the game and store it in the 'result' variable
    result = ""
    if choice == system_choice:
        print("It's a Draw")
        result = "DRAW"
    elif (choice == 1 and system_choice == 2) or (choice == 2 and system_choice == 1):
        print('Paper wins =>', end="")
        result = 'Paper'
    elif (choice == 1 and system_choice == 3) or (choice == 3 and system_choice == 1):
        print('Rock wins =>\n', end="")
        result = 'Rock'
    elif (choice == 2 and system_choice == 3) or (choice == 3 and system_choice == 2):
        print('Scissors wins =>', end="")
        result = 'Scissors'

    # Determine the winner (player, system, or a tie) and display the result
    if result == 'DRAW':
        print("<== It's a tie ==>")
    elif result == choice_name:
        print("<== User wins ==>")
    else:
        print("<== System wins ==>")

    # Ask if the player wants to play again
    print("Do you want to play again? (Yes/No)")
    ans = input().lower()

    # If the player enters 'n' or 'N', exit the loop to end the game
    if ans == 'n':
        break

# Display a thank-you message when the game ends
print("Thank you for playing")
```
